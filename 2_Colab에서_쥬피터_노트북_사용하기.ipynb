## 강의 자료가 담긴 쥬피터 노트북을 코랩에 올리기(업로드, upload)

이번 강의에 사용되는 강의 자료는 모두 쥬피터 노트북으로 작성하여 GitHub에 올려두었기 때문에 이를 내려받아서(다운로드, download) 자신의 코랩에 업로드하면 필요할 때 참고할 수 있습니다. 

1. 웹 브라우저의 주소창에 [https://github.com/joongyang/Python-for-Smart-Agriculture-Experts](https://github.com/joongyang/Python-for-Smart-Agriculture-Experts)를 입력하여 강의자료가 있는 곳에 연결합니다.

2. **Code**라는 초록색 버턴을 클릭하면 나타나는 하위 메뉴에서 제일 아래쪽에 있는 **Download ZIP**를 선택하면 전체 강의자료를 자신의 컴퓨터에 압축파일로 내려받을 수 있습니다.

3. 내려받은 압축파일의 압축을 풀면 여러 개의 쥬피터 노트북이 보일 것입니다.

4. 코랩에서 **파일** 메뉴를 클릭하면 나타나는 하위 메뉴 중에서 **노트 업로드** 메뉴를 선택하면 업로드할 파일을 선택할 수 있게 하는 창이 나타납니다.

5. 이 창의 가운데 위치한 **파일 선택** 버턴을 클릭한 다음 다운로드한 쥬피터 노트북을 선택하면 업로드가 완료되고 이들은 구글 드라이브의 **Colab Notebooks** 폴더에 저장됩니다.

6. 이제 코랩에 연결하면 언제든지 구글 드라이브의 **Colab Notebooks** 폴더에 저장된 쥬피터 노트북을 이용할 수 있습니다. 

## 쥬피터 노트북 사용하기

쥬피터 노트북을 생성하고 저장하는 방법을 알았으니 쥬피터 노트북으로 파이썬 프로그램을 작성하고 실행하는 방법에 대해 알아볼 순서입니다.

1. 쥬피터 노트북은 셀(cell)로 구성되며, 두 종류의 셀을 주로 사용합니다. 프로그래밍을 위한 **코드 셀(code cell)**과 문서 작성을 위한 **텍스트 셀(text cell)**이 그것이다.

2. `markdown`은 텍스트 셀에서 문서를 작성할 때 사용하는 언어입니다.
그러니까 텍스트 셀에는 `markdown`을 이용하여 문서를 작성합니다.
하지만 `markdown`이 수식은 지원하지 않기 때문에  수식은 `LaTeX` 양식으로 입력해야 합니다.
보기 좋고 유용한 문서를 작성하려면 `markdown`과 `LaTeX` 두 가지를 공부해야 합니다.
이는 프로그래밍과는 상관없는 것입니다.
오직 글을 쓰기 위한 것인데 데이터 분석에는 **story telling**이 매우 중요하기 때문입니다.

3. Python 프로그램은 코드 셀에 입력합니다.

4. 코드 셀에 입력한  프로그램을 실행 또는 재실행할 때는 해당 셀을 선택한 다음 `ctrl+enter`를 누르거나 셀의 왼쪽에 나타나는 실행 버턴을 누르면 됩니다.

5. 쥬피터 노트북은 하나의 노트북에 있는 모든 코드 셀에 입력된 프로그램을 하나의 프로그램으로 생각합니다.
하지만 이 프로그램을 셀 별로 실행할 수 있으며, 앞에 위치한  셀을 수정하지 않았으면 해당 셀만 실행하면 됩니다.
하지만 앞에 위치한 셀을 수정하였으면 수정된 셀 이후의 모든 셀을 다시 실행하여야 합니다.
이때 `런타임` 메뉴를 적절히 이용하면 편리합니다.

6. 셀을 삭제, 수정하거나 문서 내에서 이동할 수 있으며, 셀의 타입도 변경할 수 있습니다.
